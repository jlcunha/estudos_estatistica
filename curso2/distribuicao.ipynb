# Probabilidade

Oque é probabilidade?
É a razão entre o numero de vezes que um evento especifico ocorre e o numero total de possiveis resultados.

### Probabilidade Clássica:

Definição: Baseada na suposição de que todos os resultados possíveis de um experimento são igualmente prováveis.

Exemplo: Lançar uma moeda justa, onde há dois resultados possíveis (cara ou coroa) e ambos são considerados igualmente prováveis. Portanto, a probabilidade de obter cara ou coroa é de 1/2.


### Probabilidade Empírica (ou Frequencial):

Definição: A probabilidade empírica é baseada em observações e na frequência de ocorrência de eventos em experimentos repetidos.

Exemplo: Lançar uma moeda 100 vezes e observar que 60 vezes obteve-se cara e 40 vezes obteve-se coroa. A probabilidade empírica de obter cara é 60/100 = 0.6, e a probabilidade empírica de obter coroa é 40/100 = 0.4.

### Probabilidade Subjetiva:

Definição: A probabilidade subjetiva é baseada nas crenças ou julgamentos pessoais de um indivíduo sobre a probabilidade de um evento ocorrer. Essas crenças não precisam ser fundamentadas em observações ou dados empíricos.

Exemplo: Atribuir uma probabilidade subjetiva a um produto novo, baseado em probabilidades de produtos similares.


# <font color='yellow'> Distribuições Discretas </font>


A distribuição discreta é caracterizada por ter valores isolados e contáveis. Isso significa que os valores possíveis da variável são finitos ou infinitos, mas separados uns dos outros. Exemplos comuns de distribuições discretas incluem a distribuição de Poisson, distribuição binomial e a distribuição de Bernoulli. Essas distribuições são adequadas para modelar eventos discretos, como o número de sucessos em um número fixo de tentativas ou o tempo entre ocorrências de eventos específicos.

# <font color='LimeGreen'> Bernoulli </font>

 A distribuição de Bernoulli é uma distribuição discreta que modela um único experimento aleatório com dois resultados possíveis: sucesso ou fracasso. Ela é chamada assim em homenagem ao matemático suíço Jacob Bernoulli.

Em termos simples, a distribuição de Bernoulli é usada quando estamos interessados em eventos binários, onde o resultado pode ser representado por um valor booleano, como "sim" ou "não", "1" ou "0", "verdadeiro" ou "falso", entre outros.

Os parâmetros principais dessa distribuição são:

p (probabilidade de sucesso): representa a probabilidade de ocorrer um evento de sucesso em um único experimento.

1 - p (probabilidade de fracasso): representa a probabilidade de ocorrer um evento de fracasso em um único experimento.
A função de probabilidade de uma distribuição de Bernoulli é dada por:

### $$P(X = k) = p^k * (1 - p)^{(1 - k)}$$

onde:

- $P(X = k)$        é a probabilidade de X ser igual a k (0 ou 1).
- $p^k$             é a probabilidade de ocorrer um sucesso (k = 1).
- $(1 - p)^(1 - k)$ é a probabilidade de ocorrer um fracasso (k = 0).

#### Exemplo:

Imagine que tenhamos uma moeda viciada, em que a probabilidade de obter cara (sucesso) em um único lançamento seja de 0.3, e a probabilidade de obter coroa (falha) seja de 0.7.

Agora, vamos calcular a probabilidade de obter cara em um único lançamento dessa moeda viciada, utilizando a fórmula da distribuição de Bernoulli:

$$P(X = 1) = 0.3^1 * (1 - 0.3)^{(1 - 1)}$$
$$P(X = 1) = 0.3 * 0.7^0$$
$$P(X = 1) = 0.3 * 1$$
$$P(X = 1) = 0.3$$

```python
def bernoulli(k, p):
    """
    Aplica a formula de Bernoulli
    """
    
    termo1 = p**k
    termo2 = (1-p)**(1-k)
    prob = termo1 * termo2

    return prob

bernoulli(1, 0.3)
```

# <font color='LimeGreen'> Binomial </font>


A distribuição binomial é um modelo estatístico usado para descrever o número de sucessos em uma série de experimentos independentes e idênticos, onde cada experimento tem dois resultados possíveis: sucesso ou falha.

Os principais requisitos para que um experimento siga uma distribuição binomial são:

- Os experimentos são independentes, o que significa que o resultado de um experimento não afeta o resultado dos outros.
- Cada experimento tem apenas dois resultados possíveis: sucesso ou falha.
- A probabilidade de sucesso em cada experimento é constante e denotada por "p". A probabilidade de falha é representada por "q" e é calculada como q = 1 - p.
- O número de experimentos é fixo e é representado por "n".
- A função de probabilidade da distribuição binomial é dada pela fórmula:

### $$P(X = k) = C(n, k) * p^k * q^{(n - k)}$$

Onde:

- $P(X = k)$ é a probabilidade de obter exatamente "k" sucessos em "n" experimentos.
- $C(n, k)$ é o coeficiente binomial, que representa o número de combinações de "n" itens tomados em grupos de "k". Pode ser calculado como $C(n, k) = n! / (k! * (n - k)!)$.
- $p$ é a probabilidade de sucesso em cada experimento.
- $q$ é a probabilidade de falha em cada experimento.


### Exemplo

Imagine que estamos analisando uma campanha de marketing por e-mail, na qual enviamos mensagens promocionais para os clientes. E queremos calcular qual a probabilidade de 5 clientes comprarem o produto de e-mail.

Suponha que enviamos o e-mail promocional para 10 clientes, e que a taxa de conversão é de 0.3, qual é a probabilidade de 5 realizarem a compra.

$$P(X=5) = C(10, 5) * 0.3^{5} * 0.7^{10-5}$$
$$P(X=5) ≃ 252 * 0.00243 * 0.168$$
$$P(X=5) ≃ 0.1029$$

![image.png](https://upload.wikimedia.org/wikipedia/commons/thumb/7/75/Binomial_distribution_pmf.svg/1200px-Binomial_distribution_pmf.svg.png)

```python
import math

def binomial(k, n, p):
    """
    Aplica a formula da binomial.
    """
    termo1 = math.comb(n, k)
    termo2 = p**k
    termo3 = (1-p)**(n-k)
    prob = termo1 * termo2 * termo3

    return prob

k = 5 # numero de clientes que queremos calcular a probabilidade
n = 10 # numero total de clientes
p = 0.3 # taxa de conversão

binomial(k=k, n=n, p=p)
```

# <font color='LimeGreen'> Poisson </font>

A distribuição de Poisson é uma distribuição de probabilidade discreta que descreve a ocorrência de eventos raros e independentes em um intervalo fixo de tempo ou espaço. Ela é frequentemente usada para modelar fenômenos que envolvem contagem de eventos, como o número de chamadas recebidas em um call center, o número de acidentes em uma determinada região durante um período de tempo, ou o número de emails recebidos por hora.

A distribuição de Poisson é caracterizada por um único parâmetro, chamado de taxa média de ocorrência (λ). Esse parâmetro representa a média do número de eventos que se espera ocorrer em um determinado intervalo. A função de probabilidade da distribuição de Poisson é dada pela fórmula:

### $$P(X = k) = e^{-λ} * \frac{λ^k}{k!}$$

Onde:

- $P(X = k)$ é a probabilidade de ocorrerem exatamente k eventos,
- $e$ é a base do logaritmo natural (aproximadamente 2.71828),
- $λ$ é a taxa média de ocorrência de eventos,
- $k$ é o número de eventos que ocorreram.

A distribuição de Poisson possui algumas propriedades importantes:
- A média e a variância da distribuição são iguais e iguais a λ.
- A distribuição de Poisson se aproxima da distribuição binomial quando o número de tentativas é grande e a probabilidade de sucesso em cada tentativa é pequena.


### Exemplo
Suponha que você esteja trabalhando em uma central telefônica e deseja prever o número de chamadas recebidas em um determinado intervalo de tempo. Com base em dados históricos, você observou que, em média, ocorrem 2 chamadas por minuto.

A distribuição de Poisson pode ser usada para modelar esse cenário. Vamos supor que você queira calcular a probabilidade de receber exatamente 4 chamadas em um intervalo de 5 minutos.

$$\newcommand{\euler}{e}$$

$$P(X=4) = e^{-2}*\frac{2⁴}{4!}$$
$$P(X=4) ≃ 0.1353 * \frac{16}{24}$$
$$P(X=4) ≃ 0.1353 * 2/3$$
$$P(X=4) ≃ 0.0902$$


```python
def poisson(k, taxa_media, intervalo=1):
    """
    Aplica a formula de poison.
    """
    termo1 = math.exp(-taxa_media * intervalo)
    termo2 = ((taxa_media * intervalo) ** k)/math.factorial(k)
    prob = termo1 * termo2
    
    return prob

poisson(4, 2)
```

Vamos usar o mesmo exemplo anterior em que queremos calcular a probabilidade de receber exatamente 4 chamadas só que agora em um intervalo de 5 minutos, com uma taxa média de 2 chamadas por minuto.

```python
poisson(4, 2, 5)
```

# <font color='yellow'> Distribuições contínuas </font>

Por outro lado, a distribuição contínua é caracterizada por ter um conjunto infinito e não enumerável de valores possíveis dentro de um intervalo contínuo. A variável contínua pode assumir qualquer valor dentro desse intervalo, com uma infinidade de possibilidades. Exemplos comuns de distribuições contínuas incluem a distribuição normal (ou gaussiana), a distribuição exponencial e a distribuição uniforme. Essas distribuições são utilizadas para modelar eventos contínuos, como medidas físicas, tempo de resposta, duração de eventos, entre outros.

Uma diferença fundamental entre as duas distribuições é a forma como a probabilidade é atribuída aos valores. Na distribuição discreta, a probabilidade é atribuída a cada valor individualmente, sendo representada por uma função de probabilidade discreta. Já na distribuição contínua, a probabilidade é associada a intervalos de valores, sendo representada por uma função de densidade de probabilidade contínua.

# <font color='limegreen'> Distribuição normal </font>

A distribuição normal, também conhecida como distribuição gaussiana, é uma das distribuições de probabilidade mais importantes e amplamente utilizadas na estatística. Ela descreve uma distribuição simétrica em torno de sua média, assumindo a forma de uma curva em formato de sino.

A distribuição normal é caracterizada por dois parâmetros principais: a média (μ) e o desvio padrão (σ). A média determina o centro da distribuição, indicando onde está localizado o pico da curva. O desvio padrão controla a dispersão dos dados em relação à média, ou seja, quão espalhados estão os valores em torno da média.

A função de densidade de probabilidade (PDF) da distribuição normal é descrita pela fórmula:


## $$f(x) = \frac{1}{{\sigma \sqrt{2\pi}}} \cdot e^{-\frac{(x-\mu)^2}{2\sigma^2}}$$

Onde:

- $x$ é o valor da variável aleatória,
- $\mu$ é a média da distribuição,
- $\sigma$ é o desvio padrão da distribuição,
- $e$ é a base do logaritmo natural (constante de Euler), e
- $\pi$ é a constante matemática pi (aproximadamente 3.14159).

Essa função descreve a probabilidade de ocorrência de um determinado valor (x) em relação à média (μ) e ao desvio padrão (σ). A curva de distribuição normal é simétrica em torno da média, com uma forma característica de sino.

A distribuição normal possui várias propriedades importantes, como a área sob a curva correspondendo a uma probabilidade total de 1 e a existência de uma relação conhecida como a regra empírica, que estabelece que aproximadamente 68% dos dados estão dentro de um desvio padrão da média, 95% estão dentro de dois desvios padrão e 99.7% estão dentro de três desvios padrão.

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np

mu = 0  # Média da distribuição normal
sigma = 1  # Desvio padrão da distribuição normal

x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 100)  # Valores de x para plotagem
y = norm.pdf(x, mu, sigma)  # Função de densidade de probabilidade (PDF)


# Calculando a probabilidade de um valor menor ou igual a 1
prob = norm.cdf(1, mu, sigma)
print("Probabilidade:", prob)


plt.plot(x, y)
plt.fill_between(x, y, where=(x >= -2 * sigma) & (x <= 1.), color='blue', alpha=0.3)
plt.xlabel('Valores')
plt.ylabel('Densidade de Probabilidade')
plt.title('Distribuição Normal')
plt.grid(True)
plt.show()
```

- $\mu$ é a média da distribuição do valor

# <font color='yellow'> Teorema do limite central </font>

Suponha que tenhamos uma população qualquer com média μ e desvio padrão σ. Se tomarmos amostras aleatórias dessa população e calcularmos as médias dessas amostras, à medida que o tamanho das amostras aumenta, a distribuição das médias amostrais se aproxima cada vez mais de uma distribuição normal, independentemente da forma da distribuição original da população.

Em termos mais simples, o Teorema do Limite Central nos diz que, independentemente da forma da distribuição original, quando amostramos repetidamente e calculamos as médias, a distribuição dessas médias se aproxima de uma distribuição normal.

Existem algumas condições e suposições para a aplicação correta do Teorema do Limite Central:

As amostras devem ser selecionadas aleatoriamente da população.
O tamanho das amostras deve ser grande o suficiente (geralmente, é recomendado que seja pelo menos 30) para que a aproximação da distribuição normal seja válida.
As amostras devem ser independentes umas das outras.
O Teorema do Limite Central é muito útil na estatística, pois permite que usemos a distribuição normal para realizar inferências estatísticas, mesmo quando a distribuição da população original não é conhecida ou não é normal. Isso nos permite fazer estimativas de intervalo e realizar testes de hipóteses de forma mais confiável.

É importante ressaltar que, embora o Teorema do Limite Central seja poderoso e amplamente aplicável, existem algumas situações em que ele pode não ser adequado. Por exemplo, quando existem outliers extremos na população ou quando a dependência entre as observações é significativa, o teorema pode não se aplicar completamente.


## $$ \bar{x} \sim N(\mu, \frac{\sigma^2}{n})$$

Onde:
- $\bar{x}$: é a média amostral, que é a média dos valores observados na amostra.
- $\mu$: é a média populacional, que é o valor médio verdadeiro de toda a população.
- $\sigma^2$: é a variância populacional, que mede a dispersão dos valores em relação à média populacional.
- $n$: é o tamanho da amostra. ( cada amostra deve ter no minimo 30 unidades )

Logo, se o número de amostras for baixo, o valor da variação vai ser grande, se o **n** for aumentando a variação vai diminuindo.

Então foi se provado que isto é verdadeiro pra **n** maiores que trinta.

## $\sqrt{\frac{{\sigma^2}}{{n}}} == erro padrao$

- O valor da variação no escopo do teorema do limite central é chamado de erro padrão

```python
plt.plot(x, y)
plt.vlines(x=0, ymax=0.4, ymin=0, color='black', linestyles='--')
plt.text(x=0.1, y=0, s='$\mu$', fontsize=15)
plt.vlines(x=1, ymax=0.23, ymin=0, color='red', linestyles='--')
plt.text(x=1.1, y=0, s='$\\sqrt{\\frac{{\sigma^2}}{{n}}}$', fontsize=20)

plt.vlines(x=-1, ymax=0.23, ymin=0, color='red', linestyles='--')
plt.text(x=-2, y=0, s='$\\sqrt{\\frac{{\sigma^2}}{{n}}}$', fontsize=20)

plt.xticks([])
plt.yticks([])
plt.xlabel('erro padrão       $\mu$       erro padrão')

plt.show();
```

Na estatística, inferência refere-se ao processo de tirar conclusões ou fazer afirmações sobre uma população com base em informações limitadas contidas em uma amostra. A inferência estatística é usada quando não é viável ou prático examinar ou obter informações de toda a população, então uma amostra representativa é coletada e as conclusões são feitas com base nessa amostra.

O objetivo da inferência estatística é fornecer estimativas e tomar decisões sobre uma população com base nas informações contidas na amostra. Isso envolve a aplicação de métodos estatísticos para analisar os dados da amostra e fazer inferências sobre os parâmetros desconhecidos da população, como a média, a proporção, a variância, entre outros.

Existem duas principais abordagens na inferência estatística: a estimação de parâmetros e o teste de hipóteses.
- Estimação de parâmetros: Envolve o uso de estatísticas amostrais para estimar os valores desconhecidos dos parâmetros populacionais. Por exemplo, estimar a média populacional com base na média da amostra.

- Teste de hipóteses: É um procedimento estatístico usado para tomar decisões sobre uma afirmação (hipótese) feita sobre uma população com base em evidências fornecidas pelos dados da amostra. Envolve a formulação de uma hipótese nula (que é testada) e uma hipótese alternativa (que é considerada se a hipótese nula for rejeitada).

Resumo: Independente da distribuição que sua variavel de interesse tenha, se voce tirar varias amostras calcular a média dessas amostras e verificar a distribuição das médias, elas terão uma distribuição normal.

- O teorema do limite central permite que sejamos capazes de inferir atraves de uma amostra, se ela pertence a uma determinada população conhecida.

- Se voce tem o dominio sobre uma amostra, é possivel inferir qual é o perfil da população que ela pertence.

- Se eu tenho duas amostras, é possivel comparar-las e dizer se elas são equiparaveis.

- Se conhece uma população e se conhece uma amostra, é possivel ver se elas vem do mesmo universo.
Com uma amostra grande o suficiente pra ter uma média e um desvio padrão bem definidos. Se usa a distribuição normal como árbitro pra saber o quão distante da média populacional está a sua amostra.

- A média das médias vai ser proxima da média populacional.

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('../notebooks_data/Estatística+para+Ciência+de+Dados+e+Machine+Learning (1)/Estatística para Ciência de Dados e Machine Learning/Bases de dados/credit_data.csv')
data.head()
```

```python
def teorema_limite_central(serie, tamanho_da_amostra=30, numero_de_amostras=100):
    medias = [serie.sample(tamanho_da_amostra, random_state=i).mean() for i in range(numero_de_amostras)]
    X_barra = pd.DataFrame({'medias': medias})
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

    # Plot do gráfico da população
    ax1.hist(serie)
    ax1.set_title(f'População\nMédia: {serie.mean():.2f}')

    # Plot do gráfico da amostra
    ax2.hist(X_barra['medias'])
    ax2.set_title(f'Amostra\nMédia: {X_barra["medias"].mean():.2f}')

    plt.tight_layout()  # Para ajustar a disposição dos subplots
    plt.show()



teorema_limite_central(serie=data.age, tamanho_da_amostra=1000, numero_de_amostras=500)
```

# <font color='limegreen'> Intervalo de confiança</font>



O intervalo de confiança é uma medida estatística que fornece uma faixa de valores plausíveis para um parâmetro desconhecido de uma população com base em uma amostra. É uma estimativa da incerteza associada à estimativa pontual do parâmetro.

Em outras palavras, o intervalo de confiança nos dá uma ideia de quão precisos são nossos cálculos estatísticos. Geralmente, é expresso como um intervalo, por exemplo, "Intervalo de Confiança de 95%". Isso significa que, em média, 95% dos intervalos de confiança calculados a partir de diferentes amostras capturariam o valor real do parâmetro.

O intervalo de confiança leva em consideração o tamanho da amostra, a variabilidade dos dados e o nível de confiança desejado. Quanto maior o tamanho da amostra e menor a variabilidade dos dados, mais estreito será o intervalo de confiança. Além disso, aumentar o nível de confiança resulta em um intervalo de confiança mais amplo, pois estamos sendo mais rigorosos em nossa exigência de capturar o valor real do parâmetro.

O intervalo de confiança é uma ferramenta importante na análise estatística, pois nos permite avaliar a precisão das estimativas e fazer inferências sobre a população com base em dados amostrais.


# <font color='limegreen'> Nivel de significancia</font>



O nível de significância é um conceito estatístico utilizado em testes de hipóteses para avaliar a força das evidências contra uma hipótese nula. Representado pelo símbolo α (alfa), o nível de significância define a probabilidade máxima de rejeitar a hipótese nula quando ela é verdadeira.

Em um teste de hipóteses, estabelecemos uma hipótese nula (H0) e uma hipótese alternativa (H1). O nível de significância determina o limiar para a rejeição da hipótese nula. Se o valor-p calculado no teste for menor que o nível de significância, podemos rejeitar a hipótese nula em favor da hipótese alternativa.

O valor escolhido para o nível de significância depende da importância do teste e das consequências de rejeitar a hipótese nula incorretamente. Um nível de significância comumente utilizado é α = 0,05, o que significa que há uma probabilidade de 5% de rejeitar a hipótese nula quando ela é verdadeira. No entanto, em alguns casos, níveis mais rigorosos, como α = 0,01 ou α = 0,001, podem ser adotados para reduzir o risco de erro.

O nível de significância está relacionado ao conceito de erro tipo I em testes de hipóteses. Um erro tipo I ocorre quando rejeitamos a hipótese nula erroneamente, ou seja, concluímos que existe uma diferença ou efeito quando, na verdade, ele não existe na população. Controlar o nível de significância é fundamental para controlar o risco desse tipo de erro.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

def plot_intervalo_confianca_significancia(desvios_padroes):
    mu = 0  # Média
    sigma = 1  # Desvio padrão

    x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 100)
    y = norm.pdf(x, mu, sigma)

    fig, ax = plt.subplots(figsize=(10, 8))

    ax.plot(x, y)

    # Intervalo de confiança
    intervalo_inferior = mu - desvios_padroes * sigma
    intervalo_superior = mu + desvios_padroes * sigma
    ax.fill_between(x, y, where=((x>= intervalo_inferior) & (x <= intervalo_superior)), alpha=0.3, color='blue')

    # Nível de significância
    nivel_significancia = 1 - norm.cdf(desvios_padroes)
    intervalo_superior_significancia = mu + desvios_padroes * sigma
    intervalo_inferior_significancia = mu - desvios_padroes * sigma
    ax.fill_between(x, y, where=(x >= intervalo_superior_significancia), alpha=0.3, color='red')
    ax.fill_between(x, y, where=(x <= intervalo_inferior_significancia), alpha=0.3, color='red')

    ax.text(2.5, 0.35, r'$\frac{Nível\ de\ Significância}{2} = \frac{\alpha}{2}$', ha='center', va='center', fontsize=15)
    ax.text(-2.5, 0.35, r'$Intervalo\ de\ confiança = 1 - \alpha$', ha='center', va='center', fontsize=15)

    ax.text(intervalo_superior_significancia + 0.2, 0.05, '$\\frac{\\alpha}{2}$', ha='center', va='center', fontsize=15)
    ax.text(intervalo_inferior_significancia - 0.2, 0.05, '$\\frac{\\alpha}{2}$', ha='center', va='center', fontsize=15)

    ax.set_xlabel('Valores')
    ax.set_ylabel('Densidade de Probabilidade')
    ax.set_title(f'Intervalo de Confiança para {desvios_padroes} Erro Padrão: {(1-(2*nivel_significancia)):.4f}%\nNível de Significância:{2*nivel_significancia:.4f}%')

    plt.show()

# Exemplo de uso
desvios = 2
plot_intervalo_confianca_significancia(desvios)
```

Como o p valor esta fora do intervalo de confiança, ou seja ele é menor que o nivel de significancia, nos rejeitamos a H0
